```python
import pandas as pd

import os

import urllib.request
import numpy as np


from itertools import compress
from PIL import Image
import tempfile
from PIL import ImageFile
```

```python
## Define image reslution 

mres = 500
```

### Load the dataset

```python
flin = 'C:/Users/haozh/Desktop/buttress_root/butress_plants_iNaturalist_url_combine.csv'

dataset = pd.read_csv(flin, low_memory = False)

```

```python
taxon_id_list = dataset['taxonKey'].unique().tolist()

taxon_id_list
```

```python
# Reduce the list to the taxa that have not yet been downloaded
wi = taxon_id_list.index(4070543)
taxon_id_list = taxon_id_list[wi:]
```

```python
root_path = 'D:/Data/GBIF/'
dataset['id'] = range(len(dataset))

dataset['extension'] = dataset['pictureLinks'].apply(lambda x: os.path.splitext(x)[1])

dataset['local_path'] = dataset.apply(lambda row: f"{root_path}{row['taxonKey']}/{row['id']}{row['extension']}", axis = 1)

dataset["file_ok"] = 0

dataset['extension'].value_counts()

dataset['pictureLinks'].head()
```

```python
# =============================================================================
# # Screen trough downloaded files, fish out the broken ones, and store the names of the good ones
# =============================================================================  

ImageFile.LOAD_TRUNCATED_IMAGES = True

for taxon_id in taxon_id_list:
    
    print(taxon_id)
    
    dataset_class = dataset[dataset['taxonKey']==taxon_id]
    folder_path = 'D:/Data/GBIF/{}/'.format(taxon_id)
    
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        
    for index, row in dataset_class.iterrows():

        # check if file exists and if so, skip
        if os.path.isfile(row['local_path']):    

            # Try to open
            try: 
                image = Image.open(row['local_path'])
                continue
            except Exception as e:
                print(e)

        # Define name of tempfile
        f = tempfile.TemporaryDirectory()
        tf = f.name + '/' + os.path.basename(row['local_path'])

        try:
            # Download file
            url = row['pictureLinks']
            urllib.request.urlretrieve(url, tf) 

            # Open image
            rd = Image.open(tf)   

            # Rescale image
            if min(rd.size)>mres:
                sc_f = mres/min(rd.size)
                scim = rd.resize((round(sc_f*rd.size[0]),round(sc_f*rd.size[1])))

                # Save
                scim.save(row['local_path'])

            else:
                rd.save(row['local_path'])

        except Exception as e:
            print(e)
            print("Image not found: ", url)
        
                # check if file exists
        if os.path.isfile(row['local_path']):    

            # Try to open
            try: 
                image = Image.open(row['local_path'])
                dataset.loc[index,'file_ok'] = 1
                continue

            # If not working, remove local path name
            except Exception as e:
                print(e)
                os.remove(row["local_path"])
                dataset.loc[index,"local_path"] = ""
```
